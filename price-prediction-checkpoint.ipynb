Model for real estate

Training the Model

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

# Separate the target variable
housing = strat_train_set.drop("MEDV", axis=1)
housing_labels = strat_train_set["MEDV"].copy()

# Create the pipeline
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

# Fit the pipeline with DataFrame
my_pipeline.fit(housing)

# Transform the training data
housing_prepared = my_pipeline.transform(housing)

# Train the model
model = RandomForestRegressor()
model.fit(housing_prepared, housing_labels)
```

Saving the Model

```python
from joblib import dump, load
dump(model, "Real_estate.joblib")
dump(my_pipeline, 'pipeline.joblib')
```

Testing the Model

```python
import pandas as pd
import numpy as np
from joblib import load
from sklearn.metrics import mean_squared_error

# Load the saved pipeline and model
my_pipeline = load('pipeline.joblib')
model = load('Real_estate.joblib')

# Separate the target variable from the test set
housing_test = strat_test_set.drop("MEDV", axis=1)
housing_test_labels = strat_test_set["MEDV"].copy()

# Transform the test data using the loaded pipeline
housing_test_prepared = my_pipeline.transform(housing_test)

# Make predictions on the test data
final_predictions = model.predict(housing_test_prepared)

# Evaluate the model
final_mse = mean_squared_error(housing_test_labels, final_predictions)
final_rmse = np.sqrt(final_mse)

# Print the results
print("Final predictions:", final_predictions)
print("Actual values:", list(housing_test_labels))
print("Mean Squared Error (MSE):", final_mse)
print("Root Mean Squared Error (RMSE):", final_rmse)
```
